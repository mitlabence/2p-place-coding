```python
#import necessary packages
import pandas as pd
import h5py
import xarray as xr
import numpy as np
import scipy.sparse 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#opening the hpf5 file
hdf=h5py.File(data)
```

```python
C_data=hdf['estimates']['C']
```

```python
C=pd.DataFrame(np.array(C_data))
```

```python
#isolating twenty random cells to check the fluorescences
random_cells=np.random.choice(len(C),size=20,replace=False)

time=500 #in seconds
n_frames=int(7.5 *time) #since I am imaging for 15 X averaging * 2


#taking the first thousand values from random 20 cells
subset_C=C.iloc[random_cells,:n_frames]


#plotting the cells for specific time

#plot the cells

for i,(index,row) in enumerate(subset_C.iterrows()):
    plt.plot(np.arange(n_frames)/7.5,row+i*40)
plt.xlabel('time(s)')
plt.yticks([]) 
plt.ylabel('Fluorescent traces')
plt.title('Putative units fluorescent traces')
```
